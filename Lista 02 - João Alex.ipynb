```python
import sympy as sy
import sympy.physics.mechanics as me
from IPython.display import display,Math
from sympy.physics.vector.printing import vpprint, vlatex
```

1- A aceleração de um ponto se movendo em uma linha vertical é dada pela equação $ a = 12t -
20$. Sabe-se que o seu deslocamento é s = −10m no instante t = 0 e que o seu deslocamento é
s = 10m no instante t = 5s. Determinar a equação de seu movimento s(t).

```python
t = me.dynamicsymbols._t
a = 12*t - 20
c1 = sy.symbols('c1')
v = sy.integrate(a,t) + c1
display(Math(f'a = {sy.latex(a)}'))
display(Math(f'v = {sy.latex(v)}'))
```

```python
c2 = sy.symbols('c2')
s = sy.integrate(v,t) + c2
display(Math(f's = {sy.latex(s)}'))
```

```python
# s = −10m no instante t = 0
sy.Eq(s.subs(t,0),-10)
```

```python
s = s.subs(c2,-10)
display(Math(f's = {sy.latex(s)}'))
```

```python
# s = 10m no instante t = 5s
display(Math(f'c1 = {sy.solve(sy.Eq(s.subs(t,5), 10),c1)[0]}'))
```

```python
s = s.subs(c1,4)
display(Math('\\text{RESPOSTA:}'))
display(Math(f's = {sy.latex(s)}'))
```

2- Considerando a figura abaixo e sabendo que em um dado instante o bloco
A possui velocidade de 5m/s e aceleração de 10m/s², ambas para a esquerda,
determinar os módulos e os sentidos da velocidade e da aceleração do bloco B.



```python
x1,x2 = sy.symbols('x1,x2')
xA,yB = me.dynamicsymbols('x_A,y_B')
```

```python
# Comprimento do Cabo
L = yB + (x1 - xA) + 2*(x2 - xA)
display(Math(f'L = {sy.latex(L)}'))
```

```python
# Valocidade de Deformação do Cabo
v_L = sy.diff(L,t)
display(Math('v_L = %s '%(sy.latex(v_L))))
```

```python
# Aceleração de Deformação do Cabo
a_L = sy.diff(v_L,t)
display(Math('a_L = %s '%(sy.latex(a_L))))
```

```python
# Velocidade de A = 5m/s
v_yB = sy.solve(v_L.subs(sy.diff(xA,t),5), sy.diff(yB,t))[0]
display(Math('\\text{Velocidade do Bloco B:}'))
display(Math('v_{yB} = %s m/s' %v_yB))
```

```python
# Aceleração de A = 10m/s²
a_yB = sy.solve(a_L.subs(sy.diff(xA,t,2),10), sy.diff(yB,t,2))[0]
display(Math('\\text{Aceleração do Bloco B:}'))
display(Math('a_{yB} = %s m/s²' %a_yB))
```

3- Obtenha as expressões da velocidade absoluta e da aceleração
absoluta do ponto P, conforme mostrado na figura abaixo.



```python
h,R = sy.symbols('h,R')
l,theta,gamma = me.dynamicsymbols('l,theta,gamma')
```

```python
N = me.ReferenceFrame('N')
A = me.ReferenceFrame('A')
B = me.ReferenceFrame('B')
```

```python
A.orient(N,'Axis',[theta,N.z])
B.orient(A,'Axis',[gamma,A.x])
```

```python
O = me.Point('O')
O.set_vel(N,0)
```

```python
P = me.Point('P')
pPO = h*B.z - l*B.y
P.set_pos(O,pPO)
P.set_vel(N,pPO.dt(N))
```

```python
display(Math('v_P = %s' %vlatex(P.vel(N))))
```

```python
display(Math('v_P = %s' %vlatex(P.vel(N).express(B))))
```

```python
display(Math('a_P = %s' %vlatex(P.acc(N))))
```

4- Obtenha as expressões para a velocidade e a aceleração relativas
do ponto P em relação ao ponto Q no referencial newtoniano.



```python
R = sy.symbols('R')
theta,gamma,x = me.dynamicsymbols('theta,gamma,x')
```

```python
N = me.ReferenceFrame('N')
A = me.ReferenceFrame('A')
B = me.ReferenceFrame('B')
```

```python
A.orient(N,'Axis',[theta,N.z])
B.orient(A,'Axis',[-gamma,A.x])
```

```python
O = me.Point('O')
O.set_vel(N,0)
```

```python
Q = me.Point('Q')
pQO = R*B.y
Q.set_pos(O,pQO)
Q.set_vel(N,pQO.dt(N))
```

```python
P = me.Point('P')
pPQ = -x*B.y
P.set_pos(Q,pPQ)
P.set_vel(N,P.pos_from(O).dt(N))
```

```python
display(Math('v_{PQ} = %s ' %vlatex(P.pos_from(Q).dt(N))))
```

```python
display(Math('a_{PQ} = %s '%vlatex(P.pos_from(Q).dt(N).dt(N))))
```
